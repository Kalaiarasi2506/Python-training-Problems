```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def total(self):
        return sum(self.marks)

    def average(self):
        return sum(self.marks) / len(self.marks)

    def highest_mark(self):
        return max(self.marks)
    def status(self):
        for mark in self.marks:
            if mark < 35:
                return "Fail"
        return "Pass"
s = Student("Ravi", [78, 45, 92, 30, 67])
print("Total:", s.total())
print("Average:", s.average())
print("Highest Mark:", s.highest_mark())
print("Status:", s.status())
```

```python
class Employee:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores
    def best_score(self):
        return max(self.scores)
    def average_score(self):
        return round(sum(self.scores) / len(self.scores), 1)
    def months_above_80(self):
        count = 0
        for i in self.scores:
            if i > 80:
                count += 1
        return count
name = input("Employee Name: ")
scores = list(map(int, input("Scores: ").split()))
e = Employee(name, scores)
print("Best Score:", e.best_score())
print("Average Score:", e.average_score())
print("Months Above 80:", e.months_above_80())
```

```python
class ParkingLot:
    def __init__(self, vehicles):
        self.vehicles = vehicles
    def find_duplicates(self):
        dup = []
        for i in self.vehicles:
            if self.vehicles.count(i) > 1 and i not in dup:
                dup.append(i)
        return dup
    def unique_vehicles(self):
        return len(set(self.vehicles))
vehicles = list(map(int,input().split()))
p = ParkingLot(vehicles)
print("Duplicate Vehicles:", p.find_duplicates())
print("Unique Vehicles:", p.unique_vehicles())
```

```python
import ast
class Player:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores
    def best_score(self):
      return max(self.scores)
    def lowest_score(self):
      return min(self.scores)
    def average_score(self):
      return round(sum(self.scores) / len(self.scores), 1)
    def century_count(self):
      count = 0
      for i in self.scores:
        if i >= 50:
          count += 1
      return count
name = input("Player Name: ")
scores =ast.literal_eval(input("Scores: "))
p = Player(name, scores)
print("Best Score:", p.best_score())
print("Lowest Score:", p.lowest_score())
print("Average Score:", p.average_score())
print("Centuries:", p.century_count())
```

```python
import ast
class ShoppingCart:
    def __init__(self, items):
        self.items = items
    def total(self):
        calculated_total = 0
        for i in self.items:
            calculated_total += i
        return calculated_total
    def max_price(self):
        return max(self.items)
    def discount(self):
        current_total = self.total()
        return current_total - (current_total * 15 / 100)
items = ast.literal_eval(input("Enter prices"))
s = ShoppingCart(items)
print("Total:", s.total())
print("Max Price:", s.max_price())
print("Discounted bill:", s.discount())
```
